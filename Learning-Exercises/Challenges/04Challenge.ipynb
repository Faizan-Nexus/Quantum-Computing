# Perform the following 
- Create a three qubit entanglemnet
- The state should be $|000\rangle$ + $|111\rangle$ / 2
- Above state is also GHZ (GreenBerger-Hornie-Zilinger) state

```python
from qiskit import *
from qiskit.visualization import *
from qiskit.quantum_info import *
from qiskit_aer import *
from math import pi
```

```python
#Creating a corelated entanglement
qc = QuantumCircuit(3)
qc.data.clear()
qc.h(0)
qc.cx(0,1)
qc.ccx(0,1,2)
qc.measure_all()
qc.draw('mpl')
```

```python
backend  = Aer.get_backend('qasm_simulator')
check = transpile(qc,backend)
job = backend.run(check, shots =1000)
result = job.result()
```

```python
plot_histogram(result.get_counts())
```
