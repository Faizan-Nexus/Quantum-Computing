```python
from qiskit import *
from qiskit.visualization import plot_histogram
from qiskit_aer import Aer
```

```python
#Creating a classical and Quantum Register
q = QuantumRegister(1,'q')
c = ClassicalRegister(1,'c')
#Creating a Quantum Circuit q is quatum qubit and c is classical bit where the result of this quantum circuit will be stored
qc = QuantumCircuit(q,c)
```

```python
#Performing Operations and drawing the final ciruit
qc.h(q)
qc.measure(q,c)
qc.draw('mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
compiled = transpile(qc,backend=simulator)
job = simulator.run(compiled, shots= 1024)
result = job.result()
plot_histogram(result.get_counts())
```
