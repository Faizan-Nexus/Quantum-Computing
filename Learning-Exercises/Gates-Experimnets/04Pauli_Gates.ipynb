```python
from qiskit import *
from qiskit.visualization import *
from qiskit.quantum_info import *
from qiskit_aer import *
from math import pi
```

# Pauli X-Gate

```python
qc1 = QuantumCircuit(1)
qc1.data.clear()
#This will put our qubit in superposition
qc1.ry(pi/4,0)
qc1.x(0)
# qc.x(0)
qc1.draw('mpl')
```

```python
backend  = Aer.get_backend('qasm_simulator')
job = backend.run((transpile(qc1,backend)),1024).result
state1 = Statevector.from_instruction(qc1)

array_to_latex(state1, prefix = "\\text {statevector =} \n")
```

```python
plot_bloch_multivector(state1)
```
