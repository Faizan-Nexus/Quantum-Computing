```python
from qiskit import *
from qiskit.visualization import *
from qiskit.quantum_info import *
from qiskit_aer import *
from math import pi
```

```python
#Creating a corelated entanglement
qc = QuantumCircuit(2)
qc.data.clear()
qc.h(0)
qc.cx(0,1)
qc.measure_all()
qc.draw('mpl')
```

```python
backend  = Aer.get_backend('qasm_simulator')
check = transpile(qc,backend)
job = backend.run(check, shots =1000)
result = job.result()
```

```python
plot_histogram(result.get_counts())
```

```python
#Creating the anti-related Qunatum States
qc1 = QuantumCircuit(2)
qc1.data.clear()
qc1.h(0)
qc1.x(1)
qc1.cx(0,1)
qc1.measure_all()
qc1.draw('mpl')
```

```python
result = backend.run(transpile(qc1,backend),shots = 1000).result()
```

```python
plot_histogram(result.get_counts())
```
