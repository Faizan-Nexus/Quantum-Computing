```python
from qiskit import *
from qiskit.quantum_info import *
from qiskit_aer import *
from qiskit.visualization import *
from math import pi
```

```python
qc = QuantumCircuit(3)
qc.draw('mpl')
```

```python
#Choosing the backend and executing the ciruit on backend
back = Aer.get_backend('qasm_simulator')
job = back.run(transpile(qc,back),1024).result
```

```python
state = Statevector.from_instruction(qc)
plot_bloch_multivector(state)
```

```python
qc.rx(pi/3,0)
qc.ry(pi/3,1)
qc.h(2)         #To expilicilty notice the effect of Rz we apply the hadmard to put the bit in superposition
qc.rz(pi/3,2)
qc.draw('mpl')
```

```python
job = back.run(transpile(qc,back),1024).result
state = Statevector.from_instruction(qc)
plot_bloch_multivector(state)
```

```python
#We can alos manipulate them as following
qc.data.clear()
qc.rx(2,0)
qc.ry(-pi/3,1)
qc.h(2)         #To expilicilty notice the effect of Rz we apply the hadmard to put the bit in superposition
qc.rz(pi/3,2)
qc.draw('mpl')
```

```python
job = back.run(transpile(qc,back),1024).result
state = Statevector.from_instruction(qc)
plot_bloch_multivector(state)
```
