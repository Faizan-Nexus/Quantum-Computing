# CNOT Gate

```python
from qiskit import *
from qiskit.quantum_info import *
from qiskit_aer import *
from qiskit.visualization import *
from math import pi
```

```python
qc = QuantumCircuit(3)
qc.data.clear()
qc.x(0)
qc.cx(0,1)
qc.cx(1,2)
qc.draw('mpl')
```

```python
#Choosing the backend and executing the ciruit on backend
back = Aer.get_backend('qasm_simulator')
job = back.run(transpile(qc,back),1024).result
```

```python
state = Statevector.from_instruction(qc)
plot_bloch_multivector(state)
```

# Toffoli Gate (CCNOT)

```python
circuit = QuantumCircuit(3)
circuit.data.clear()
circuit.x([0,1])
circuit.rx(pi/3,2)    #How it's rotation works
circuit.draw('mpl')
```

```python
check = transpile(circuit,back)
job = back.run(check,shots= 10000)
result = job.result()
#StateVector
state1 = Statevector.from_instruction(circuit)
plot_bloch_multivector(state1)
```

```python
# Applying toffoli to see the rotation effect
circuit.ccx(0,1,2)
circuit.draw('mpl')
```

```python
check = transpile(circuit,back)
job = back.run(check,shots= 10000)
result = job.result()
#StateVector
state1 = Statevector.from_instruction(circuit)
plot_bloch_multivector(state1)
```

## Swap Gate
- This gate Swaps the states of two qubits
## Controlled Swap (Fredkin Gate)
- This gate perfroms swap operation but when control qubit is $|1\rangle$

```python
qcircuit = QuantumCircuit(3)
qcircuit.data.clear()
qcircuit.h(1)
qcircuit.rx(-pi/3,2)
qcircuit.draw('mpl')
```

```python
check = transpile(qcircuit,back)
job = back.run(check,shots= 10000)
result = job.result()
#StateVector
state1 = Statevector.from_instruction(qcircuit)
plot_bloch_multivector(state1)
```

```python
#Applying Swap Gate
qcircuit.swap(1,2)
qcircuit.draw('mpl')
```

```python
check = transpile(qcircuit,back)
job = back.run(check,shots= 10000)
result = job.result()
#StateVector
state1 = Statevector.from_instruction(qcircuit)
plot_bloch_multivector(state1)
```

```python
# Modifying the Circuit
qcircuit = QuantumCircuit(3)
qcircuit.data.clear()
qcircuit.x(0)
qcircuit.h(1)
qcircuit.rx(-pi/3,2)
qcircuit.draw('mpl')
```

```python
#Statevecotr Before Applying Fredkin
check = transpile(qcircuit,back)
job = back.run(check,shots= 10000)
result = job.result()
#StateVector
state1 = Statevector.from_instruction(qcircuit)
plot_bloch_multivector(state1)
```

```python
#Applying CSwap/Fredkin Gate
qcircuit.cswap(0,1,2)
qcircuit.draw('mpl')
```

```python
check = transpile(qcircuit,back)
job = back.run(check,shots= 10000)
result = job.result()
#StateVector After Applying the swap gate
state1 = Statevector.from_instruction(qcircuit)
plot_bloch_multivector(state1)
```
